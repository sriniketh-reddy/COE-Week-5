```python
data = {'Difficulty': ['Easy','Medium','Hard','Medium','Easy'],
       'Test Score':[50,60,70,80,60]}
df = pd.DataFrame(data)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["encoded_Difficulty"]=le.fit_transform(df["Difficulty"])
```

```python
df["encoded_Difficulty"].corr(df["Test Score"])
```

```python
df.select_dtypes(include="number").corr()
```

```python
data1 = {'Color': ['Red', 'Blue', 'Green', 'Blue', 'Red'],
        'Test Score': [50, 60, 70, 80, 90]}
df1 = pd.DataFrame(data1)
df1_encoded = pd.get_dummies(df1, columns=['Color'])
print(df1_encoded)
correlation = df1_encoded['Color_Red'].corr(df1_encoded['Test Score'])
print("Correlation with Red:", correlation)
```
