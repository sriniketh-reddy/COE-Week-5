# 1️⃣ Handling Missing Values

#### Check for NaN (missing values) in all columns

```python
df.info()
```

#### Fill missing values using mode (for categorical) and median (for numeric) where appropriate.


```python
num_cols = df.select_dtypes(include=["number"]).columns
cat_cols = df.select_dtypes(exclude=["number","bool"]).columns
print(num_cols,cat_cols)
```

###### Finding the % of unique values in Categorical columns

```python
percent_of_unique_vals_in_cat_cols = []
for i in cat_cols:
    percent_of_unique_vals_in_cat_cols.append((i,len(df[i].unique())/(len(df)-df[i].isna().sum())*100))
percent_of_unique_vals_in_cat_cols
```

```python
less_than_50_percent = [col[0] for col in percent_of_unique_vals_in_cat_cols if col[1]<50]
less_than_50_percent
```

```python
for col in less_than_50_percent:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.info()
```

```python
#<sender>_Dataset_<datasetname>
cols_lesser_than_70_percent_values = []
len_of_df = len(df)
for col in df.columns:
    if (len_of_df-df[col].isna().sum())/len_of_df<0.70:
        cols_lesser_than_70_percent_values.append(col)
df = df.drop(columns=cols_lesser_than_70_percent_values+["Notes"])
```

# 2️⃣String Operations & Text Processing

#### Clean JobDescription and JobRequirment by:
`Removing HTML tags, special characters, and excessive whitespace.`

```python
import re
```

```python
def remove_noise_from_text(text):
    text = re.sub(r'<.*?>', '', str(text))
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text


df["JobDescription"] = df["JobDescription"].apply(remove_noise_from_text)
df["JobRequirment"] = df["JobRequirment"].apply(remove_noise_from_text)
```
